#### 5.2.2 Generating trajectories

Let's create a random walk for 500 steps.

```python
num_steps = 500

x_step = list()
for x in rng(num_steps):
    if x < 0.5:
        x_step.append(1)
    else:
        x_step.append(-1)
        
y_step = list()
for x in rng(num_steps):
    if x < 0.5:
        y_step.append(1)
    else:
        y_step.append(-1)
        
random_walk = np.array([x_step, y_step])
random_walk_new = random_walk.cumsum(1)

plt.plot(random_walk_new[0,], random_walk_new[1,], '-o');
```

#### 5.8 Numerical solution of differential equations

```python
from scipy.integrate import odeint
```

```python
# simple_oscillator
def F(y, t):
    dy = [0, 0]
    dy[0] = y[1]
    dy[1] = -y[0]
    return dy
```

```python
# solve ode
t_min = 0
t_max = 10
dt = 0.1
t = np.arange(t_min, t_max+dt, dt)

initial_conditions = [(1.0, 0.0), (0.0, 1.0)]

plt.figure()
for y0 in initial_conditions:
    y = odeint(F, y0, t)
    plt.plot(t, y[:, 0], linewidth = 2)
    
skip = 5
t_test = t[::skip]
plt.plot(t_test, np.cos(t_test), 'bo')
plt.plot(t_test, np.sin(t_test), 'go');
```

#### 6 Generate two-dimensional random walks

```python
from numpy.random import random as rng
num_steps = 1000

x_step = list()
for x in rng(num_steps):
    if x < 0.5:
        x_step.append(1)
    else:
        x_step.append(-1)
        
y_step = list()
for x in rng(num_steps):
    if x < 0.5:
        y_step.append(1)
    else:
        y_step.append(-1)
        
random_walk = np.array([x_step, y_step])
random_walk_new = random_walk.cumsum(1)

plt.plot(random_walk_new[0,], random_walk_new[1,], '-o')
plt.axis('equal');
```

```python
plt.figure()

for i in range(4):
    num_steps = 1000
    x_step = list()
    for x in rng(num_steps):
        if x < 0.5:
            x_step.append(1)
        else:
            x_step.append(-1)
        
    y_step = list()
    for x in rng(num_steps):
        if x < 0.5:
            y_step.append(1)
        else:
            y_step.append(-1) 
        
    random_walk = np.array([x_step, y_step])
    random_walk_new = random_walk.cumsum(1)
    plt.subplot(2, 2, i+1)
    plt.plot(random_walk_new[0,], random_walk_new[1,], '-o')
    plt.axis('equal');
```

#### 7.1 Image processing

A digital image is a collection of pixels.

In the case of a black and white photograph, each pixel is represented by a number corresponding to the intensity (mean photon rate) sensed at a point in the xy plane of the detector.

A color image consists of three such arrays, reporting the intensities of red, blue and green light.

A computer uses the pixel data in an image file to reproduce the image on your monitor.

As far as Python is concerned, then, every image is an array of numbers.
Conversely, any array with the appropriate shape and data content can be displayed as an image.
This mapping between arrays and images allows us to use Python to import, analyze, transform, and save images.

```python
# Read an image from a file into an array.
photo = plt.imread('lena.PNG')
```
